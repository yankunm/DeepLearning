## Series

Two strcutures in pandas: 

* Series - indexed array
* DataFrame - 2D table

Both datastructures canhold any type of data in python

We will need to learn indexing, axis labeling, and alignment. Data alignment is automatic.

```python
data = [1, 2, 3]
s = pd.Series(data)
print(s)
```

```python
sims = {"sim0": 1, "sim1":2}
sims = pd.Series(sims, index=['sim0', 'b', 'sim1'])

# sim0 got pulled out, while the other indexes were not found,
# NaN is pandas marker, not a number
```

```python
# iloc is the slicing operation here 
sims.iloc[:3]
```
