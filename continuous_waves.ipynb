```python
cw = td.ContinuousWave(freq0=2e14, fwidth=20e12)
runtime = 1e-13

source = td.PlaneWave(
    source_time = cw,
    size=(td.inf, td.inf, 0),
    center=(0, 0, 0),
    direction="-",
    pol_angle=0
)

plot_source(source, runtime, save_as="start.png")
```

```python
def plot_source(source, run_time, sampling=5001, freqs=None, xlim_time=None, xlim_freq=None, save_as=None):
    f, (ax1, ax2) = plt.subplots(1, 2, tight_layout=True, figsize=(10, 4))
    times=np.linspace(0, run_time, sampling)
    ax1 = source.source_time.plot(times, ax=ax1)
    ax1.set_ylim(-1, 1)
    if xlim_time is not None: ax1.set_xlim(xlim_time)

    ax2 = source.source_time.plot_spectrum(times, val='abs', ax=ax2)
    ax2.set_ylim(0, 2e-14)
    if freqs is not None: ax2.fill_between([np.min(freqs), np.max(freqs)], [-0e-16, -0e-16], [60e-16, 60e-16], alpha=0.4, color='g')
    if xlim_freq is not None: ax2.set_xlim(xlim_freq)
    if save_as is not None: plt.savefig(save_as, dpi=300)
    plt.show()


runtimes = [1e-15, 1e-14, 1e-13, 5e-13, 1e-12, 1e-11, 1e-10, 1e-9, 1e-8]

cw = td.ContinuousWave(freq0=2e14, fwidth=50e12)
source = td.PlaneWave(
    source_time = cw,
    size=(td.inf, td.inf, 0),
    center=(0, 0, 0),
    direction="-",
    pol_angle=0
)

def different_runtimes(runtimes):
    for runtime in runtimes: 
        print(f"{runtime:.1e} seconds")
        plot_source(source, runtime, save_as=f"plots/f{runtime:.1e}plot.png")

different_runtimes(runtimes)
```

```python
runtimes = [5e-13]
fwidth = [1e11, 1e12, 10e12, 20e12, 30e12, 40e12, 50e12, 60e12, 70e12, 80e12, 90e12]

def plot_source(source, run_time, sampling=5001, freqs=None, xlim_time=None, xlim_freq=None, save_as=None):
    f, (ax1, ax2) = plt.subplots(1, 2, tight_layout=True, figsize=(10, 4))
    times=np.linspace(0, run_time, sampling)
    ax1 = source.source_time.plot(times, ax=ax1)
    ax1.set_ylim(-1, 1)
    if xlim_time is not None: ax1.set_xlim(xlim_time)

    ax2 = source.source_time.plot_spectrum(times, val='abs', ax=ax2)
    ax2.set_ylim(0, 2e-14)
    if freqs is not None: ax2.fill_between([np.min(freqs), np.max(freqs)], [-0e-16, -0e-16], [60e-16, 60e-16], alpha=0.4, color='g')
    if xlim_freq is not None: ax2.set_xlim(xlim_freq)
    if save_as is not None: plt.savefig(save_as, dpi=300)
    plt.show()


def different_fwidth(fwidth):
    for freqw in fwidth:
        cw = td.ContinuousWave(freq0=2e14, fwidth=freqw)
        source = td.PlaneWave(
            source_time = cw,
            size=(td.inf, td.inf, 0),
            center=(0, 0, 0),
            direction="-",
            pol_angle=0
        )
        plot_source(source, runtime, xlim_freq=(0, 5e14), save_as=f"plots/freqw{freqw:.1e}.png")

different_fwidth(fwidth)
```
